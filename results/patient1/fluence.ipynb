````python
from IPython.display import Image
import matplotlib.pyplot as plt
import numpy as np
````

I accessed what looks like the fluence maps using the following:
```python
import connect
beam_set = connect.get_current('BeamSet')
for beam in beam_set.Beams:
    for segment in beam.Segments:
        fluence_map = segment.Fluence.FluenceData
```

So to get the `jj`th fluence map from the ``ii``th beam, use
```python
fluence_map = beam_set.Beams[ii].Segments[jj].Fluence.FluenceData
```
