# Initial Results

RayStation results from changing the weight, dose, and volume objective function parameters for the lungs.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Changing OAR Weight Parameter

```python
# Load results
weights = np.load('lung_weights.npy')
lung_avg_w = np.load('lung_avgs_w.npy')
ptv_d95_w = np.load('ptv_vals_w.npy')
```

```python
# OAR Average Dose
plt.plot(weights, lung_avg_w, 'o')
plt.xscale('log')
plt.xlabel('OAR Weight Parameter')
plt.ylabel('OAR Average Dose')
```

```python
# PTV D95 Dose
plt.plot(weights, ptv_d95_w, 'o')
plt.xscale('log')
plt.xlabel('OAR Weight Parameter')
plt.ylabel('PTV D95 Dose')
```

```python
# Both
plt.plot(lung_avg_w, ptv_d95_w, 'o')
plt.xlabel('OAR Average Dose')
plt.ylabel('PTV D95 Dose')
```

## Changing OAR Dose Parameter

```python
# Load results
doses = np.load('lung_doses.npy')
lung_avg_d = np.load('lung_avgs_d.npy')
ptv_d95_d = np.load('ptv_vals_d.npy')
```

```python
# OAR Average Dose
plt.plot(doses, lung_avg_d, 'o')
plt.xlabel('OAR Dose Parameter')
plt.ylabel('OAR Average Dose')
```

```python
# PTV D95 Dose
plt.plot(doses, ptv_d95_d, 'o')
plt.xlabel('OAR Dose Parameter')
plt.ylabel('PTV D95 Dose')
```

```python
# Both
plt.plot(lung_avg_d, ptv_d95_d, 'o')
plt.xlabel('OAR Average Dose')
plt.ylabel('PTV D95 Dose')
```

## Changing OAR Volume Parameter

```python
# Load results
volumes = np.load('lung_volumes.npy')
lung_avg_v = np.load('lung_avgs_v.npy')
ptv_d95_v = np.load('ptv_vals_v.npy')
```

```python
# OAR Average Dose
plt.plot(volumes, lung_avg_v, 'o')
plt.xlabel('OAR Volume Parameter')
plt.ylabel('OAR Average Dose')
```

```python
# PTV D95 Dose
plt.plot(volumes, ptv_d95_v, 'o')
plt.xlabel('OAR Volume Parameter')
plt.ylabel('PTV D95 Dose')
```

```python
# Both
plt.plot(lung_avg_v, ptv_d95_v, 'o')
plt.xlabel('OAR Average Dose')
plt.ylabel('PTV D95 Dose')
```

### All Together

```python
plt.plot(lung_avg_w, ptv_d95_w, 'o')
plt.plot(lung_avg_d, ptv_d95_d, 'o')
plt.plot(lung_avg_v, ptv_d95_v, 'o')
plt.xlabel('OAR Average Dose')
plt.ylabel('PTV D95 Dose')
plt.legend(['Weights', 'Doses', 'Volumes'])
```
